```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
from sklearn import metrics
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
df = pd.read_csv("USA_Housing.csv")
```

```python
df.head() # avg area show city location of house
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df["Price"])
```

```python
# df.corr()
sns.heatmap(df.corr(), annot=True)
```

```python
df.columns
```

```python
# Features
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
# Label, what we are trying to predict
y = df["Price"]
```

```python
# Now split the data into train and test data
# test_size = % of test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
# X, y = np.arange(10).reshape((5, 2)), range(5)
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
# create a linear regression object
lm = LinearRegression()
```

```python
# train or fit my model for (traning data)
lm.fit(X_train, y_train)
```

```python
print(lm.intercept_) # grab the intercept
```

```python
# relat to each feature of dataset
lm.coef_
```

```python
X_train.columns
```

```python
cdf = pd.DataFrame(data=lm.coef_, index=X.columns, columns=["Coeff"]) # 1 unit increase in price related to feature
```

```python
cdf
```

```python
# Predictions on the dataset only features
# predicted prices of the house
predictions = lm.predict(X_test) # model never seen before
```

```python
# predictions
```

```python
# y_test correct prices of the house
```

```python
plt.scatter(y_test, predictions)
```

```python
# difference b/w the actual value while test and the predict value
plt.figure(figsize=(12, 5))
sns.distplot((y_test - predictions))
```

```python
print(f'Mean absolute error (MAE): {metrics.mean_absolute_error(y_test, predictions)}')
print(f'Mean absolute error (MSE): {metrics.mean_squared_error(y_test, predictions)}')
print(f'Mean absolute error (RMSE): {np.sqrt(metrics.mean_squared_error(y_test, predictions))}')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
print(lmm.intercept_)
```

```python
print(lmm.coef_)
```

```python
cdf = pd.DataFrame(data=lmm.coef_, index=X.columns, columns=["Coeff"])
```
