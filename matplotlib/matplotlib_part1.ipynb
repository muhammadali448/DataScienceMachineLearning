```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

```python
# Functional
plt.plot(x, y, "r-")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
plt.subplot(1, 4, 1)
plt.plot(x, y, "r")
plt.subplot(1, 4, 2)
plt.plot(x, y, "b")
plt.subplot(1, 4, 3)
plt.plot(x, y, "g")
plt.subplot(1, 4, 4)
plt.plot(x, y, "g")
```

```python
# object oriented method
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y, "r")
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_title("Title")
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.2])
ax1.plot(x, y, "r")
ax1.set_title("Larger")
ax2.plot(x, y, "b")
ax2.set_title("Smaller")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# for ax in axes:
#     ax.plot(x, y)
axes[0].plot(x, y, "r")
axes[0].set_title("Left")
axes[1].plot(x, y, "g")
axes[1].set_title("Right")
# plt.tight_layout()
```

```python
# Figsize and Dpi
fig = plt.figure(figsize=(6, 6))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
# Figsize and Dpi
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))
axes[0].plot(x, y, "r")
axes[0].set_title("Left")
axes[1].plot(x, y, "g")
axes[1].set_title("Right")
plt.tight_layout()
```
