```python
import requests
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import pandas.plotting._converter as pandacnv
pandacnv.register() # just for future warning
```

```python
def get_live_prices(symbol, convert=["USD"]):
    url = f"https://min-api.cryptocompare.com/data/price?fsym={symbol.upper()}&tsyms={','.join(convert).upper()}"
    return requests.get(url).json()
```

```python
def get_prices_minute(symbol, limit, aggregate, convert=["USD"]):
    url = f"https://min-api.cryptocompare.com/data/histominute?fsym={symbol}&tsym={','.join(convert).upper()}&limit={limit}&aggregate={aggregate}"
    response = requests.get(url).json()
    df = pd.DataFrame(response["Data"])
    df["timestamp"] = [datetime.datetime.fromtimestamp(d) for d in df.time]
    return df
```

```python
# 1 minute
time_delta = 1 
df = get_prices_minute(symbol="BTC", limit=100, aggregate=time_delta)
df_1 = get_prices_minute(symbol="ETH", limit=100, aggregate=time_delta)
fig, axes = plt.subplots(figsize=(15,5), nrows=1, ncols=2)
for ax in axes:
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
axes[0].plot(df.timestamp, df.close)
axes[0].set_xticklabels(df.timestamp.dt.strftime('%Y-%m'), rotation=45)
axes[0].set_title("Bitcoin")
axes[1].plot(df_1.timestamp, df_1.close, 'r')
axes[1].set_title("Ethereum")
axes[1].set_xticklabels(df_1.timestamp.dt.strftime('%Y-%m'), rotation=45)

fig
plt.tight_layout()
```
