```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

```python
# Functional
plt.plot(x, y, "r-")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, y, "r")
plt.subplot(1, 3, 2)
plt.plot(x, y, "b")
plt.subplot(1, 3, 3)
plt.plot(x, y, "g")
```

```python
# object oriented method
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x, y, "r")
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_title("Title")
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.2])
ax1.plot(x, y, "r")
ax1.set_title("Larger")
ax2.plot(x, y, "b")
ax2.set_title("Smaller")
```
