```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data[0].shape
```

```python
# plt.scatter(x=data[0][:,0], y=data[0][:,1])
# 1 is distinct and other are noisy
sns.scatterplot(x=data[0][:,0], y=data[0][:,1], hue=data[1], palette="rainbow")
```

```python
data[1] # correct data
```

```python
# Unsupervised
kmeans = KMeans(n_clusters=8)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_ # predicted labels
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(12, 6))
ax1.set_title("K Means")
sns.scatterplot(x=data[0][:,0], y=data[0][:,1], 
                hue=kmeans.labels_, palette="rainbow", ax=ax1)
# ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap="rainbow")
ax2.set_title("Original")
# ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")
sns.scatterplot(x=data[0][:,0], y=data[0][:,1], 
                hue=data[1], palette="rainbow", ax=ax2)
# which point belong which group set
# their is some noise in original dataset
```
