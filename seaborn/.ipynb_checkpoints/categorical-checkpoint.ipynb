```python
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set_style('darkgrid')
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# sns.barplot(x="sex", y="total_bill", data=tips)
# sns.barplot(x="sex", y="total_bill", hue="day", data=tips)
sns.barplot(x="sex", y="total_bill", hue="day", data=tips, estimator=np.std) # x is categorical and y is some numeric
```

```python
sns.countplot(x="sex", data=tips, hue="day")
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker", palette="Set2")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="smoker", palette="Set2", split=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue="sex", dodge=True)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color="black")
```

```python
sns.pointplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
sns.boxenplot(x="day", y="total_bill", data=tips, hue="smoker", palette="Set2")
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", data=tips, kind="box") # for all plots defined in kind
```
