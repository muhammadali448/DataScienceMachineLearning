```python
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set_style('darkgrid')
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=["o", "v"]
          , scatter_kws={ "s": 100 })
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="day", hue="sex", aspect=0.6, height=6)
```
